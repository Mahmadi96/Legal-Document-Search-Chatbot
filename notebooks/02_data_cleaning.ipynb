# 02 - Data Cleaning and Validation

Checks missing and duplicate values, performs second-pass extraction for parties and legal articles, and validates the resulting WTO case-law dataset.

**Expected input:** `data/processed/wto_cases.csv` and the source PDFs  
**Generated output:** updated `data/processed/wto_cases.csv`

> Install dependencies once from the repository root with `pip install -r requirements.txt`. Run the notebooks in numerical order.

```python
from pathlib import Path

# Resolve repository paths whether Jupyter starts in the repository root or notebooks/
PROJECT_ROOT = Path.cwd().resolve()
if PROJECT_ROOT.name == "notebooks":
    PROJECT_ROOT = PROJECT_ROOT.parent

DATA_DIR = PROJECT_ROOT / "data"
RAW_PDF_DIR = DATA_DIR / "raw" / "wto_case_pdfs"
PROCESSED_DIR = DATA_DIR / "processed"
RESULTS_DIR = PROJECT_ROOT / "results"

RAW_PDF_DIR.mkdir(parents=True, exist_ok=True)
PROCESSED_DIR.mkdir(parents=True, exist_ok=True)
RESULTS_DIR.mkdir(parents=True, exist_ok=True)

RAW_DATA_PATH = PROCESSED_DIR / "wto_cases.csv"
PROCESSED_DATA_PATH = PROCESSED_DIR / "preprocessed_wto_cases.csv"

print(f"Project root: {PROJECT_ROOT}")
```

```python
import pandas as pd

# Finding blank and duplicate "Case IDs"

# Load the dataset
csv_path = RAW_DATA_PATH
df = pd.read_csv(csv_path)

# Find blank (empty or missing) Case IDs
blank_case_ids = df[df["Case ID"].isna() | (df["Case ID"].str.strip() == "")]
print("✅ Blank Case IDs Found:")
print(blank_case_ids)

# Find duplicate Case IDs
duplicate_case_ids = df[df.duplicated(subset=["Case ID"], keep=False)]
print("\n✅ Duplicate Case IDs Found:")
print(duplicate_case_ids)
```

```python
import pdfplumber
import pandas as pd
import os

# Second round of data extraction for "Complainant & Respondent" 

# Define the paths
folder_path = RAW_PDF_DIR
csv_path = RAW_DATA_PATH

# Load the cleaned dataset
df = pd.read_csv(csv_path)

# Ensure filenames match Case IDs in dataset
pdf_files = {f.replace("sum_e", "").replace(".pdf", "").strip(): f for f in os.listdir(folder_path) if f.endswith(".pdf")}

# Function to extract Complainant and Respondent from the PARTIES column in the blue table
def extract_parties_from_table(tables):
    complainant, respondent = None, None

    for row in tables:
        if row and len(row) >= 2:  # Ensure at least two columns exist
            # Extract Complainant (everything under "Complainant" in PARTIES column)
            if row[0] and "Complainant" in row[0]:
                complainant = row[1].strip() if row[1] else None

            # Extract Respondent (everything under "Respondent" in PARTIES column)
            if row[0] and "Respondent" in row[0]:
                respondent = row[1].strip() if row[1] else None

    return complainant, respondent

# Lists to store extracted data
complainant_data = []
respondent_data = []
error_log = []

# Loop through dataset rows, extracting from corresponding PDFs
for case_id in df["Case ID"]:
    pdf_path = os.path.join(folder_path, pdf_files.get(case_id, ""))

    try:
        with pdfplumber.open(pdf_path) as pdf:
            first_page = pdf.pages[0] if pdf.pages else None

            if first_page:
                tables = first_page.extract_table()
                
                if tables:
                    complainant, respondent = extract_parties_from_table(tables)
                    complainant_data.append(complainant if complainant else "Not Found")
                    respondent_data.append(respondent if respondent else "Not Found")
                else:
                    complainant_data.append("Not Found")
                    respondent_data.append("Not Found")
                    error_log.append(f"❌ No table found in {pdf_path}")
            else:
                complainant_data.append("Not Found")
                respondent_data.append("Not Found")
                error_log.append(f"❌ No pages found in {pdf_path}")

    except Exception as e:
        print(f"❌ Error processing {pdf_path}: {e}")
        complainant_data.append("Error")
        respondent_data.append("Error")
        error_log.append(f"❌ Error processing {pdf_path}: {e}")

# Add extracted data to dataset
df["Complainant"] = complainant_data
df["Respondent"] = respondent_data

# Save updated dataset
updated_csv_path = RAW_DATA_PATH
df.to_csv(updated_csv_path, index=False)

print(f"✅ Complainant and Respondent extracted and saved at: {updated_csv_path}")
```

```python
import pdfplumber
import pandas as pd
import os

# Second round of data extraction for "Articles" 

# Paths
folder_path = RAW_PDF_DIR
csv_path = RAW_DATA_PATH

# Load the existing dataset
df = pd.read_csv(csv_path)

# Ensure filenames match Case IDs in dataset
pdf_files = {f.replace("sum_e", "").replace(".pdf", "").strip(): f for f in os.listdir(folder_path) if f.endswith(".pdf")}

# Function to extract Articles from "AGREEMENT" or "AGREEMENTS" column
def extract_articles_from_table(tables):
    agreement_index = None
    agreements_index = None
    extracted_articles = None

    # Locate the correct column for AGREEMENT(S)
    header_row = tables[0]  # Assume first row contains column names
    for i, col_name in enumerate(header_row):
        if col_name:
            if "AGREEMENT" in col_name.upper():  
                agreement_index = i
            if "AGREEMENTS" in col_name.upper():  
                agreements_index = i

    # Extract Articles: First check "AGREEMENT", then "AGREEMENTS"
    for row in tables[1:]:  # Skip header
        if agreement_index is not None and len(row) > agreement_index and row[agreement_index]:
            extracted_articles = row[agreement_index].strip()
            break  # Take first valid entry
        elif agreements_index is not None and len(row) > agreements_index and row[agreements_index]:
            extracted_articles = row[agreements_index].strip()
            break

    return extracted_articles

# List to store updated Articles
updated_articles = df["Articles"].tolist()  # Copy current Articles column
error_log = []

# Loop through dataset rows, updating missing Articles values
for i, row in df.iterrows():
    if pd.isna(row["Articles"]) or row["Articles"].strip() == "":  # If missing
        case_id = row["Case ID"]
        pdf_path = os.path.join(folder_path, pdf_files.get(case_id, ""))

        try:
            with pdfplumber.open(pdf_path) as pdf:
                first_page = pdf.pages[0] if pdf.pages else None

                if first_page:
                    tables = first_page.extract_table()

                    if tables:
                        extracted_articles = extract_articles_from_table(tables)
                        if extracted_articles:
                            updated_articles[i] = extracted_articles
                        else:
                            error_log.append(f"❌ Articles missing in {pdf_path}")
                    else:
                        error_log.append(f"❌ No table found in {pdf_path}")
                else:
                    error_log.append(f"❌ No pages found in {pdf_path}")

        except Exception as e:
            print(f"❌ Error processing {pdf_path}: {e}")
            error_log.append(f"❌ Error processing {pdf_path}: {e}")


# Update the dataset with the extracted Articles
df["Articles"] = updated_articles

# Save updated dataset
updated_csv_path = RAW_DATA_PATH
df.to_csv(updated_csv_path, index=False)

print(f"✅ Missing Articles extracted and saved at: {updated_csv_path}")
```
